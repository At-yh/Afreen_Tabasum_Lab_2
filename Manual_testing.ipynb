```python
# Test code for circle and rectangle classes

from circle import Circle
from rectangle import Rectangle 
import math
```

```python
# Helper to make titles
def show(title):
    print("\n" + "=" * len(title))
    print(title)
    print("=" * len(title))
```

```python
print("Equality Tests")

c1 = Circle(0, 0, radius=5)
c2 = Circle(10, 10, radius=5)    # same radius -> equal
c3 = Circle(0, 0, radius=8)         # different radius -> not equal

# Rectangles: same width + height
r1 = Rectangle(width=3, height=4)
r2 = Rectangle(width=3, height=4)

print("c1 == c2 (same radius):", c1 == c2)
print("c1 == c3 (different radius):", c1 == c3)
print("r1 == r2 (same size):", r1 == r2)

print("Circle == Rectangle:", c1 == r1)   # Shapes of different types are never equal
```

```python
print("Area / Perimeter / Translate")

c = Circle(0, 0, radius=1)
print("Circle:", c)
print("Area:", c.area)
print("Perimeter:", c.perimeter)
print("Is unit circle:", Circle.is_unit_circle())
```

```python
# Move circle
c.translate(3, 4)
print("Circle after move:", c)
```

```python
# Rectangle ex
rect = Rectangle(width=2, height=5)
print("\nRectangle:", rect)
print("Area:", rect.area)
print("Perimeter:", rect.perimeter)
```

```python
# Move rectangle
rect.translate(-1, -2)
print("Rectangle after move:", rect)
```

```python
# Comparing shapes by area
print("Area Comparision Tests")

big_circle = Circle(0, 0, radius=3)
small_rect = Rectangle(width=2, height=2)

print("big_circle area:", big_circle.area)
print("small_rect area:", small_rect.area)
```

```python
# Comparision results
print("big_circle > small_rect:", big_circle > small_rect)
print("big_circle >= small_rect:", big_circle >= small_rect)
print("small_rect < big_circle:", small_rect < big_circle)
```

```python
# Error handling
try:          # radius cannot be negative
    Circle(0, 0, radius=-1)
except Exception as e:
    print("Negative radius error:", type(e).__name__, "-", e)
```

```python
# Width must be a number
try: 
    Rectangle(width="hello", height=5)
except Exception as e:
    print("Invalid width error:", type(e).__name__, "-", e)
```

```python
# cannot compare shape with number
try:
    c = Circle(0,0, radius=2)
    result = c > 100  # invalid
except Exception as e:
    print("Invalid compare error:", type(e).__name__, "-", e)
```
