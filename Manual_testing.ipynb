```python
# Test code for circle and rectangle classes

from circle import Circle
from rectangle import Rectangle 
```

```python
# Create some circles

circle1 = Circle(x=0, y=0, radius=1)   #  unit circle
cicle2 = Circle(x=1, y=1, radius=1)
circle3 = Circle(radius=3)    # default center (0,0)
```

```python
# Create rectangles

rect1 = Rectangle(x=0, y=0, width=1, height=1)
rect2 = Rectangle(x=2, y=3, width=2, height=4)
```

```python
# Equality tests

print("circle1 == circle2:", circle1 == circle2)  # True (same radius)
print("circle2 == rect1:", circle2 == rect1)      # False (different type)
print("rect1 == rect2:", rect1 == rect2)          # False (different sizes)
```

```python
#  Area and perimeter

print("circle3 area:", circle3.area)
print("circle3 perimeter:", circle3.perimeter)
print("rect2 area:", rect2.area)
print("rect2 perimeter:", rect2.perimeter)
```

```python
#  Comparison tests

print("circle3 > circle1:", circle3 > circle1)
print("rect2 >= rect1:", rect2 >= rect1)
print("circle1 < rect1:", circle1 < rect1)
```

```python
#  Translate shapes

print("Before translate:", circle1)
circle1.translate(5, 3)
print("After translate:", circle1)
```

```python
#  Error handling

try:
    circle1.translate("THREE", 5)
except TypeError as e:
    print("Error caught", e)

try:
    bad_rect = Rectangle(width=3, height="5")
except TypeError as e:
    print("Error caught", e)
```

```python
#  Special checks

print("circle1 is unit circle:", circle1.is_unit_circle())
print("circle2 is unit circle:", circle2.is_unit_circle())
print("rect1 is square:", rect1.is_square())
print("rect2 is square:", rect2.is_square())
```
