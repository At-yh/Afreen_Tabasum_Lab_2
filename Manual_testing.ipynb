```python
print("Equality Tests")

c1 = Circle(0, 0, radius=5)
c2 = Circle(10, 10, radius=5)    # same radius -> equal
c3 = Circle(radius=8)         # different radius -> not equal

# Rectangles: same width + height
r1 = Rectangle(width=3, height=4)
r2 = Rectangle(width=3, height=4)

print("c1 == c2 (same radius):", c1 == c2)
print("c1 == c3 (different radius):", c1 == c3)
print("r1 == r2 (same size):", r1 == r2)

print("Circle == Rectangle:", c1 == r1)   # Shapes of different types are never equal
```

```python
print("Area / Perimeter / Translate")

c = Circle(0, 0, radius=1)
print("Circle:", c)
print("Area:", c.area)
print("Perimeter:", c.perimeter)
print("Is unit circle:", c.is_unit_circle())
```

```python
# Move circle
c.translate(3, 4)
print("Circle after move:", c)
```

```python
# Comparing shapes by area
print("Area Comparision Tests")

big_circle = Circle(radius=3)
small_rect = Rectangle(width=2, height=2)

print("big_circle area:", big_circle.area)
print("small_rect area:", small_rect.area)
```

```python
# Error handling
class Shape:
    def translate(self, dx: float, dy: float) -> None:
        if not self._is_number(dx) or not self._is_number(dy):
            raise TypeError("translate dx and dy must be numbers (int or float)")
        self._x += float(dx)
        self._y += float(dy)
```

```python
#  Special checks
def is_unit_circle(self) -> bool:
    return math.isclose(self.radius, 1.0)
```
